```python
%matplotlib inline
import dill
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = 'p_value_results.pkl'
with open(filename, 'rb') as f:    
    results = dill.load(f)
```

```python
fig, axes = plt.subplots(2,3)
strings = ['VI', 'IS', 'PSIS']
axes_1 = axes[0,:]
axes_2 = axes[1,:]

for i, ax in enumerate(axes_2):
    ax.hist(results[i]['sigma'], bins = 10, density=True, label="Sigma", color = "turquoise", ec='blue')
    ax.plot([0,1], [1,1], '--', c='r')
    ax.set_title("$\sigma$ - %s (p=%s)" % (strings[i], round(stats.kstest(results[i]['sigma'], cdf='uniform').pvalue, 2)))
    
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
                 ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(16)

        
for i, ax in enumerate(axes_1):
    ax.hist(results[i]['beta_age'], bins = 10, density=True, label="Sigma", color = "turquoise", ec='blue')
    ax.plot([0,1], [1,1], '--', c='r')
    ax.set_title("Age - %s (p=%s)" % (strings[i], round(stats.kstest(results[i]['beta_age'], cdf='uniform').pvalue, 2)))
    
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
                 ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(16)

fig.set_size_inches(16,10)
plt.savefig("PSIS_VSBC.png")
plt.show()
```
